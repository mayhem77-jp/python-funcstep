```python
def greet(name):
    # 関数greetを定義。nameは呼び出し側から受け取る引数。
    return f"こんにちは、{name}さん！"
    # returnで「処理結果」を呼び出し元に返す。

# 関数を呼び出して、戻り値をmsgに代入。
msg = greet("masaki")

# msgの中身を表示（関数の外で実行される）
print(msg)
```
