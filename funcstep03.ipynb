```python
#リストを処理する関数
def count_long_words(words):
    #関数名count_long_wordsを定義
    #for（ループ文）のカウンタをリセット
    count=0
    #wordsというリストから、wに1つずつ入れていく
    for w in words:
    #もしwが5文字以上なら
        if len(w)>=5:
    #count+は1
            count+=1
    ♯繰り返し
    return count


#呼び出し
words=["apple","cat","banana","dog","python"]
print(count_long_words(words))
#for（ループ文）
#1回目→w="apple"
#2回目→w="cat"
#3回目→w="banana"... "python”まで繰り返す
```
