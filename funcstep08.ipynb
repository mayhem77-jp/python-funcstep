```python
#数字を入力して判定する関数
def number(n):
    if n == 0:
        return "Zero"
    elif n < 0:
        return "Negative"
    else:
        return "Positive"

# ユーザー入力
n = int(input("数字を入力してください: "))
print(number(n))

#input は関数の外で受け取る
#int() で数値に変換する
#関数は「判定だけ」担当する

```
