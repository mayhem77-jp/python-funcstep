```python
# リストの中の数字を2倍にして新しいリストとして返す関数
def double_list(numbers):
    # 空のリストを用意（ここに2倍した値を追加していく）
    result = []

    # numbers の中身を n に1つずつ取り出して処理する
    for n in numbers:
        # n を2倍にして result に追加する
        result.append(n * 2)

    # 最終的に完成した result を返す
    return result

# 関数を呼び出して結果を表示
print(double_list([1, 3, 5]))
```

```python
# リストの中の数字を2倍にして新しいリストとして返す関数
def double_list(numbers):
    # 内包表記：
    # [n * 2 for n in numbers]
    #
    # 意味を分解するとこうなる：
    # 1. for n in numbers → numbers の中身を n に1つずつ取り出す
    # 2. n * 2 → 取り出した n を2倍にする
    # 3. [] の中に書くことで「処理結果をリストとして集める」
    #
    # つまり「numbers をループしながら、2倍した値を新しいリストにして返す」という意味
    return [n * 2 for n in numbers]

# 呼び出し
print(double_list([1, 3, 5]))
```
