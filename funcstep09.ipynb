```python
#step09 のゴール
#ユーザーが 何回でも数字を入力できる
#q を入力したら終了
#入力された数字を関数に渡して判定
#結果を毎回表示する
#これができたら、「関数を使い回す」っていう本来の使い方が体で理解できる。

#複数の入力をループで処理する(07,08,09と進化)
#数値を受け取って、正、負、ゼロを返す関数(step07の正解)
def number(n):
    if n == 0:
        return "Zero"
    elif n < 0:
        return "Negative"
    else:
        return "Positive"

#while文（条件付き反復）
#与えられた条件が満たされている間、インデントが一つ下がった箇所のコードを繰り返し実行する。
#while True:で無限ループとなる（breakでループから抜ける）
while True:
    user = input("数字を入力してください（終了するには q ）: ")

#もしユーザーがｑを押せば終了
    if user == "q":
        print("終了します")
#whileの無限ループがこのBreak で終了
        break

#try文：危ない処理を入れる箱
# （n=int(user)が実行されたとき、整数以外の入力があるかもしれない。
    try:
#入力された数字を文字列⇒整数に変換（ここで、関数"number"の引数"n"の値がきまる）
        n = int(user)
#判定関数（number(n))を呼び出し
        print(number(n))
#expect節：その処理が失敗したときの逃げ道
    except ValueError:
        print("数字を入力してください")
```
