```python
#文字列を受け取って、その文字列の長さと、最初の文字を返す関数
#まず関数設定
def analyze(text):
    #文字列の長さを測る
    length=len(text)
    #最初の文字をとる
    first=text[0]
    #返す値をカンマで並べるとタプルになる
    return length,first

#呼び出し
result=analyze("Python")
print(result)

 #タプルではなく、リストにしたいならreturn[a,b]と書く



    
```

```python
# 文字列を受け取って、その長さと最初の文字を返す関数
def analyze(text):
    # 文字列の長さを取得
    length = len(text)

    # 最初の文字を取得（インデックス0）
    first = text[0]

    # カンマで並べるとタプルとして返る
    return length, first

# 呼び出し
result = analyze("Python")
print(result)

# ※ リストで返したい場合は return [length, first] と書く
#Python では「= の前後にスペース」が推奨されてる。
```
