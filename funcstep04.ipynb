```python
# タプルを受け取って、名前・年齢・都市を表示する関数
def show_person(person):
    # タプルをアンパックして3つの変数に分ける
    name, age, city = person

    # フォーマットして表示
    print(f"{name}({age}) - {city}")

# 呼び出し
p = ("masaki", 42, "Osaka")
show_person(p)
```
