```python
#数値を受け取って、正、負、ゼロを返す関数
def number(n):
    if n == 0:
        print("Zero!")
    elif n < 0:
        print("Negative!")
    else:
        print("Positiove!")

#呼び出し
num=number(-5)

#改善する点
#print ではなく return を使う方が「関数らしい」
#スペルミス：Positive → Positiove
#呼び出し側で結果を受け取るなら print(num) が必要
#今は num = number(0) してるけど、関数が print してるから num には何も入らない（None）。
```

```python
# 数値を受け取って、正・負・ゼロを返す関数（道場主ver.）
def number(n):
    if n == 0:
        return "Zero"
    elif n < 0:
        return "Negative"
    else:
        return "Positive"

# 呼び出し
result = number(0)
print(result)
```
